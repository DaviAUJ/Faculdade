```python
import numpy as np
from skimage.io import imread
import matplotlib.pyplot as plt
```

```python
def printIntervalo(imagem: np.ndarray) -> None:
    min = imagem.min()
    max = imagem.max()
    
    print(f"[{min} ({type(min).__name__}), {max} ({type(max).__name__})]")
```

# QUESTÃO 1
Esta é a imagem original que peguei do arquivo

```python
carro_original = imread("HI00008918.png", as_gray=True)
plt.imshow(carro_original, cmap="gray")
```

É possível ver que a escala de cinza da imagem está contida no [0.0, 1.0]. Isto será alterado depois, como requisitado no enunciado da questão

```python
printIntervalo(carro_original)
```

Aqui temos a função que escolhi entre aquelas apresentadas

Calculei o inverso e apliquei cada um dos gammas em cópias diferentes

```python
gamma1 = 2
gamma2 = 1 / gamma1
```

```python
carro_gamma1 = carro_original.copy() ** gamma1
plt.imshow(carro_gamma1, cmap="gray")
```

```python
carro_gamma2 = carro_original.copy() ** gamma2
plt.imshow(carro_gamma2, cmap="gray")
```

Com as duas copias feitas e funções aplicadas, agora é hora de mudar o intervalo das imagens para uint8

```python
carro_convertido1 = carro_gamma1.copy()
carro_convertido1 *= 255
printIntervalo(carro_convertido1)
```

```python
carro_convertido2 = carro_gamma2.copy()
carro_convertido2 *= 255
printIntervalo(carro_convertido2)
```

```python
carro_convertido1 = carro_convertido1.astype('uint8')
printIntervalo(carro_convertido1)
```

```python
carro_convertido2 = carro_convertido2.astype('uint8')
printIntervalo(carro_convertido2)
```

Agora com a conversão feita vou exibí-las mais uma vez

```python
plt.imshow(carro_convertido1, cmap="gray")
```

```python
plt.imshow(carro_convertido2, cmap="gray")
```

Agora vou gerar os histogramas e mostrá-los aqui

```python
h1, _ = np.histogram(carro_convertido1, bins=256)
plt.plot(h1, '-k')
```

```python
h2, _ = np.histogram(carro_convertido2, bins=256)
plt.plot(h2, '-k')
```

# QUESTÃO 2

Para fazer a equalização usarei uma função que eu mesmo fiz para aplicar o processamento na imagem da Lena

Aqui em baixo estão a imagem original, seu intervalo da escala de cinza e histograma

```python
lena_original = imread("lena_gray.bmp")
plt.imshow(lena_original, cmap="gray")
```

```python
printIntervalo(lena_original)
```

```python
h, _ = np.histogram(lena_original, bins=256)
plt.plot(h, '-k')
```

Aqui está a função que criei, preferi fazer de um jeito mais tradicional pois não conheço a sintaxe bizarra do python.
Primeiro faço uma cópia para depois fazer as operações e retornar no fim.
Depois faço a soma de prefixo que uso como tabela para fazer o cálculo dos novos valores de cinza

Fontes que usei para entender o algoritmo:
- https://pt.wikipedia.org/wiki/Equaliza%C3%A7%C3%A3o_de_histograma
- https://medium.com/@rajilini/histogram-equalization-all-you-want-to-know-13b0884c5296

```python
def equalizacao(imagem: np.ndarray) -> np.ndarray:
    out = imagem.copy()
    h, _ = np.histogram(out, bins=256)
    dim = out.shape
    
    for i in range(1, h.size):
        h[i] += h[i - 1]

    for i in range(0, dim[0]):
        for j in range(0, dim[1]):
            out[i][j] = np.round((h[out[i][j]] * 255) / h[-1])
            
    return out
```

Aqui está a imagem equalizada junto do intervalo de cinza e seu histograma

```python
lena_equalizada = equalizacao(lena_original)
plt.imshow(lena_equalizada, cmap="gray")
```

```python
printIntervalo(lena_equalizada)
```

```python
h, _ = np.histogram(lena_equalizada, bins=256)
plt.plot(h, '-k')
```
