### Quantizao em 1 bit

```python
rose_bw = rose.copy()
rose_bw[rose_bw > 127] = 255
rose_bw[rose_bw < 255] = 0
```

```python
plt.imshow (rose_bw, cmap='gray')
```

```python
linha=rose_bw[600,:]
plt.plot (linha,'-k')
```

### Amostragem 64x64

```python
rose_sa = rose.copy()
N=16
for i in range (0,1024,N):
    for j in range (0,1024,N):
        rose_sa[i:i+N, j:j+N] = rose[i,j]
```

```python
plt.imshow(rose_sa,cmap='gray')
```
