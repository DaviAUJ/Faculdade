## Operação tophat
Esta operação é usada para remover áreas homogêneas de uma imagem. Nas regiões não homogêneas, prevalesce os níveis de cinza mais claros (top). As áreas homogêneas também podem ser eliminadas por uma operação morfológica similar que, ao invés de manter os valores de nível de cinza mais altos nas partes não homogêneas, mantém os mais baixos. A operação equivalente chama-se bottom-hat.

```python
import numpy as np
from skimage.morphology import white_tophat as tophat
from skimage.morphology import black_tophat as bothat
from skimage.morphology import square, dilation, erosion, binary_dilation, binary_erosion
from skimage.io import imread
from skimage.filters import sobel
import matplotlib.pyplot as plt
```

O exemplo a seguir é usado para destacar a placa de um veículo de todo o resto ao seu redor. A placa não é uma região homogêneas e tem uma forma retangular, que pode ser explorada pela geometria de EE definidos nas operações tophat, abertura e fechamento.

```python
img = (imread('image_0016.jpg',as_gray=True)*255).astype('uint8')
plt.imshow(img,cmap='gray')
```

## Dilatação

```python
result = dilation (img, square (3))
plt.imshow (result,cmap='gray')
```

## Erosão

```python
result = erosion (img, square (3))
plt.imshow (result,cmap='gray')
```

### Suavização

```python
abertura = dilation (erosion (img, square(3)), square(3))
result = erosion (dilation (abertura, square(3)), square(3))
```

```python
plt.imshow (result, cmap='gray')
```

### Gradiente morfológico

```python
dilatacao = dilation (img, square(3))
erosao = erosion (img, square(3))
gradiente = dilatacao - erosao
```

```python
plt.imshow (gradiente, cmap='gray')
```

## Abertura e Fechamento

### tophat
A forma do EE importa. Neste caso, a operação tophat é baseada no EE quadrado de lado 11.

```python
# Abertura
ee = square (3)
abertura = erosion ((dilation (img,ee)),ee) 
fechamento = dilation ((erosion (img,ee)),ee) 
_,ax = plt.subplots (1,2)
ax[0].imshow (abertura,cmap='gray')
ax[1].imshow (fechamento,cmap='gray')
```

```python
t=tophat(img,square(11))
plt.imshow(t,cmap='gray')
aux = t.copy()
```

### Dilatação
A dilatação com EE quadrado de lado 3 irá destruir as inscrições da placa. Como o EE é pequeno, a placa será preservada mas as inscrições serão destruídas.

```python
ee=np.ones((3,3))
t=dilation(aux,ee)
plt.imshow(t,cmap='gray')
# t=sobel(t)
```

### Erosão
A erosão com um EE de geometria similar, mas menor do que a placa, irá destruir ou atenuar a maior parte do que foi destacado no passo anterior mas irá preservar a placa.

```python
ee=np.ones((11,22))
t=erosion(t,ee)
plt.imshow(t,cmap='gray')
```

```python
ee = np.ones ((5,40))
result = erosion (dilation (t,ee),ee)
plt.imshow (result,cmap='gray')
```

```python
limiar = (result.max() + result.min())/2
aux = result.copy()
aux[aux<limiar]=0
aux[aux>0]=1
plt.imshow (aux * img,cmap='gray')
```
