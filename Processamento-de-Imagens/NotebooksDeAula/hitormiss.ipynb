```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from skimage.morphology import binary_erosion, binary_dilation
```

```python
a = (1-io.imread('a.pbm')/255).astype('uint8')
x = (1-io.imread('x.pbm')/255).astype('uint8')
w = (1-io.imread('w.pbm')/255).astype('uint8')
print('A=',a)
print('X=',x)
print('W=',w)
```

```python
wx=np.zeros((7,7)).astype('uint8')
wx[1:6,1:6]=x
wx=w-wx
print('W-X=',wx)
plt.imshow(wx,cmap='gray')
```

```python
ac=1-a
print('Ac=',ac)
plt.imshow(ac,cmap='gray')
```

```python
p1=binary_erosion(a,x)
print('A erode X=',p1.astype('uint8'))
plt.imshow(p1,cmap='gray')
```

```python
p2=binary_erosion(ac,wx)
print('Ac erode (W-X)=',p2.astype('uint8'))
plt.imshow(p2,cmap='gray')
```

```python
p3=p1*p2
print('A hitormiss (X,W)=',p3.astype('uint8'))
plt.imshow(p3,cmap='gray')
```
