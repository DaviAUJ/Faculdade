```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from skimage.morphology import binary_erosion, binary_dilation
```

## Abertura e Fechamento (Sec. 9.3 do livro texto)

### Abertura 

```python
img = (io.imread ('squares.tif')/255).astype('int')  # realiza a leitura na forma de imagem binaria
```

```python
plt.imshow (img,cmap='gray')
```

```python
ee = np.ones ((13,13))
```

```python
result = binary_erosion (img, ee)
```

```python
result.dtype
```

```python
plt.imshow (result,cmap='gray')
```

```python
result = binary_dilation (result, ee)
```

```python
plt.imshow (result,cmap='gray')
```
