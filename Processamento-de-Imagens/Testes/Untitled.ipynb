```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
from scipy.signal import convolve2d
```

```python
konata_original = imread("konata.jpeg")
lena_original   = imread("../NotebooksDeAula/lena_gray.bmp")
plt.imshow(konata_original)
```

```python
konata_cinza = konata_original[:, :, 0]
plt.imshow(konata_cinza, cmap="gray")
```

```python
filtro = np.asarray([
    [1, 1, 1],
    [1, -8, 1],
    [1, 1, 1]
])

imagem = lena_original
filtro_inv = -filtro
filtrado0 = convolve2d(imagem, filtro, "same", "wrap")
filtrado1 = convolve2d(imagem, filtro_inv, "same", "wrap")

_, ax = plt.subplots(1, 2)
ax[0].imshow(filtrado0, cmap="gray")
ax[1].imshow(filtrado1, cmap="gray")
```
